Numpy数组（ndarray）的核心特性
多维性：支持各维度
同质性：同一个ndarray必须数据类型一致
高效性：基于连续内存块存储支持向量化运算

## 多维性

```python
import numpy as np
```

```python
arr0 = np.array(5)
print(arr0)
print('arr的维度:',arr0.ndim)  #打印arr的维度 number of dimensions
```

```python
arr1 = np.array([1, 2, 3])
print(arr1)
print('arr的维度:',arr1.ndim)
```

```python
arr2 = np.array([[1, 2, 3], [4, 5, 6]])
print(arr2)
print('arr的维度:',arr2.ndim)
```

## 同质性

```python
# 会自动把数据类型转化为一样的
arr4 = np.array([1, 'hello', 3])
print(arr4)
```

```python
# 会自动把数据类型转化为一样的
arr4 = np.array([1, 2.5, 3])
print(arr4)
```

```python
arr = np.array([1,2,3])
print(arr)
print(arr.shape)
print(arr.size)
print(arr.dtype)
```

```python
arr2 = np.array([[1, 2, 3], [4, 5, 6]])
print(arr2)
print(arr2.shape)
print(arr2.size)
print(arr2.dtype)
```

```python
arr2 = np.array([[1, 2, 3], [4, 5, 6],[9.3, 4, 9]])
print(arr2)
print(arr2.shape)
print(arr2.size)
print(arr2.dtype)
print(arr2.T)
```
