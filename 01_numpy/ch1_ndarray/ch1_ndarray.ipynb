Numpy数组（ndarray）的核心特性
多维性：支持各维度
同质性：同一个ndarray必须数据类型一致
高效性：基于连续内存块存储支持向量化运算

## 多维性

```python
import numpy as np
```

```python
arr0 = np.array(5)
print(arr0)
print('arr的维度:',arr0.ndim)  #打印arr的维度 number of dimensions
```

```python
arr1 = np.array([1, 2, 3])
print(arr1)
print('arr的维度:',arr1.ndim)
```

```python
arr2 = np.array([[1, 2, 3], [4, 5, 6]])
print(arr2)
print('arr的维度:',arr2.ndim)
```

## 同质性

```python
# 会自动把数据类型转化为一样的
arr4 = np.array([1, 'hello', 3])
print(arr4)
```

```python
# 会自动把数据类型转化为一样的
arr4 = np.array([1, 2.5, 3])
print(arr4)
```

### 基础构造

```python
# 使用列表的构造
list1 = [4, 5, 6]
arr = np.array(list1, dtype=np.float64)
print(arr.ndim)
print(arr.shape)
print(arr)

# 使用copy会复制一份出来
arr1 = np.copy(arr)
print(arr1)
arr1[0] = 6
print(arr1)
print(arr)
```

## 预定义形状填充

```python
# 全0数组
arr = np.zeros((2,3))
print(arr)
print(arr.dtype)

#全1数组
arr = np.ones((5,6),dtype=int)
print(arr)
print(arr.dtype)
print(arr.shape)

# 未初始化，随机数
arr = np.empty((3,5))
print(arr)
print(arr.dtype)
print(arr.shape)

# 用指定数字填满
arr = np.full((3,4), 2026)
print(arr)
print(arr.dtype)
print(arr.shape)
```

```python
# 复制形状数字填充
arr1 = np.zeros_like(arr)
print(arr1)
arr1 = np.ones_like(arr)
print(arr1)
arr1 = np.empty_like(arr)
print(arr1)
arr1 = np.full_like(arr,2027)
print(arr1)
```

## 等差数列与等间隔序列

```python
arr = np.arange(2,11,2) # 等同于range
print(arr)

arr = np.linspace(1,10,5) # 1到10取等间隔
print(arr)
arr = np.linspace(0,100,5) # 1到100取等间隔
print(arr)


# 对数间隔数列
arr = np.logspace(0 ,10, 25, base = 2) # 默认底数时10， 取等距离的数出来取对数
print(arr)
```

## 特殊矩阵的构造


```python
# 构建单位矩阵
arr = np.eye(3, 4, dtype=int)
print(arr)

#构建对角矩阵
arr = np.diag([5,1,2,3])  # 写主对角线的数字
print(arr)
```

## 随机数组的生成

```python
# 随机数组的生成
# 生成0到1随机浮点数（均匀分布）
arr = np.random.rand(2,3) #2 3 为生成数组形状
print(arr)

arr = np.random.randint(3, 6, (2, 3))
print(arr)

arr = np.random.uniform(3, 6, (2, 3))
print(arr)
```

```python
# 符合正态分布的随机数列
arr = np.random.randn(2, 3)
print(arr)
```

```python
# 设置随机种子

np.random.seed(20)
arr = np.random.randint(1,10,(2,5))
print(arr)
```
