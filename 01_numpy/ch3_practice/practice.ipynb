某城市一周最高气温[28,30,29,31,32,30,29]
1. 计算平均气温，最高气温和最低气温
2. 找出气温超过30的天数

```python
import numpy as np
```

```python
arr = np.array([28,30,29,31,32,30,29])
max_day = np.argmax(arr)
min_day = np.argmin(arr)
avg_num = np.mean(arr)
print(arr[max_day],' ',arr[min_day],' ',avg_num)
print(len(arr[arr>30]))
print(np.sum(np.where(arr > 30,1,0)))
```

生成随机数（0，10）
矩阵（3，4）
- 计算每列最大值每行最小值
- 将数组的所有奇数替换为 -1

```python
arr = np.random.randint(0, 10, size=(3, 4))
print(arr)
for i in range(4):
    print(np.max(arr[:,i]))
for i in range(3):
    print(np.min(arr[i,:]))
# 行列的最值
print(np.max(arr,axis=0))  # axis 0为行 1为列
print(np.min(arr,axis=1))
print(np.where(arr%2 != 0,-1,arr))
# 关于替换
arr[arr%2!=0] = -1
print(arr)
```
