# 第 2 章：Series

## 2.1 创建 Series

Series = **一维数组 + 索引（Index）**，每个元素都有一个对应的标签。

````python
# ① 从列表创建（默认整数索引 0, 1, 2...）
s1 = pd.Series([10, 20, 30, 40])
print("① 从列表创建:")
print(s1)

# ② 自定义索引
s2 = pd.Series([85, 92, 78, 96], index=['语文', '数学', '英语', '物理'], name='小明成绩')
print("\n② 自定义索引:")
print(s2)

# ③ 从字典创建（键 → 索引，值 → 数据）
s3 = pd.Series({'北京': 2154, '上海': 2487, '广州': 1867, '深圳': 1756})
print("\n③ 从字典创建（万人口）:")
print(s3)

# ④ 从 NumPy 数组创建
s4 = pd.Series(np.linspace(0, 1, 5), name='归一化')
print("\n④ 从 ndarray 创建:")
print(s4)

# ⑤ 标量广播：整个 Series 填充同一值
s5 = pd.Series(0, index=['a', 'b', 'c', 'd'])
print("\n⑤ 标量广播:")
print(s5)
````

## 2.2 Series 核心属性

````python
s = pd.Series([85, 92, 78, 96, 88], index=['语文','数学','英语','物理','化学'], name='小明成绩')

print(f".values  → {s.values}")          # NumPy 数组
print(f".index   → {list(s.index)}")     # 索引
print(f".dtype   → {s.dtype}")           # 数据类型
print(f".name    → {s.name}")            # 名称
print(f".shape   → {s.shape}")           # 形状 (n,)
print(f".size    → {s.size}")            # 元素总数
print(f".ndim    → {s.ndim}")            # 维度（始终为 1）
print(f".is_unique → {s.is_unique}")     # 值是否唯一
````

## 2.3 Series 索引与切片

| 方式 | 语法 | 说明 |
|------|------|------|
| 标签索引 | `s['语文']` | 按标签名取值 |
| 位置索引 | `s.iloc[0]` | 按整数位置取值 |
| 标签切片 | `s['语文':'数学']` | **包含**右端点 |
| 位置切片 | `s.iloc[0:2]` | **不包含**右端点 |
| 布尔索引 | `s[s > 85]` | 条件筛选 |

````python
print("单个标签:", s['数学'])
print("多个标签:", s[['语文', '英语', '化学']].values)
print("标签切片（含右端）:", s['语文':'英语'].values)

print("\n位置索引 s.iloc[1]:", s.iloc[1])
print("位置切片 s.iloc[1:3]:", s.iloc[1:3].values)
print("倒序 s.iloc[::-1]:", s.iloc[::-1].values)

print("\n布尔索引 s[s >= 90]:")
print(s[s >= 90])

# isin —— 多值匹配
print("\nisin 筛选:")
print(s[s.index.isin(['数学', '物理'])])
````

## 2.4 Series 运算与常用方法

````python
s_a = pd.Series({'语文': 85, '数学': 92, '英语': 78}, name='小明')
s_b = pd.Series({'数学': 88, '英语': 95, '历史': 82}, name='小红')

# 运算时按索引自动对齐，缺失标签产生 NaN
print("s_a + s_b（自动对齐）:")
print(s_a + s_b)

# 用 add 的 fill_value 参数填补缺失
print("\ns_a.add(s_b, fill_value=0):")
print(s_a.add(s_b, fill_value=0))

print("\n--- 常用统计方法 ---")
print(f"sum()    = {s.sum()}")
print(f"mean()   = {s.mean():.2f}")
print(f"max()    = {s.max()}")
print(f"min()    = {s.min()}")
print(f"std()    = {s.std():.2f}")
print(f"median() = {s.median()}")
print(f"idxmax() = {s.idxmax()}")   # 最大值的标签
print(f"idxmin() = {s.idxmin()}")   # 最小值的标签
print(f"rank():\n{s.rank()}")       # 排名

print("\n--- 其他常用方法 ---")
print(f"sort_values():\n{s.sort_values()}")
print(f"\ndescribe():\n{s.describe()}")
````

# 第 3 章：DataFrame

## 3.1 创建 DataFrame

DataFrame = **带行索引（index）和列名（columns）的二维表格**，每一列是一个 Series。

````python
# ① 从字典（列名 → 列数据）
df1 = pd.DataFrame({
    '姓名': ['小明', '小红', '小华', '小丽'],
    '年龄': [18, 19, 20, 18],
    '成绩': [85, 92, 78, 96],
    '城市': ['北京', '上海', '广州', '北京']
})
print("① 字典创建:")
print(df1)

# ② 从嵌套列表（行方向）
df2 = pd.DataFrame(
    [['小明', 18, 85], ['小红', 19, 92]],
    columns=['姓名', '年龄', '成绩'],
    index=['row1', 'row2']
)
print("\n② 嵌套列表创建:")
print(df2)

# ③ 从字典列表（每个字典是一行）
records = [
    {'产品': 'A', '价格': 100, '销量': 50},
    {'产品': 'B', '价格': 200, '销量': 30},
    {'产品': 'C', '价格': 150},         # 缺少销量 → NaN
]
df3 = pd.DataFrame(records)
print("\n③ 字典列表创建（含缺失值）:")
print(df3)

# ④ 从 NumPy 数组
df4 = pd.DataFrame(np.random.randint(60, 100, (3, 4)),
                   index=['小明', '小红', '小华'],
                   columns=['语文', '数学', '英语', '物理'])
print("\n④ NumPy 数组创建:")
print(df4)
````

## 3.2 DataFrame 核心属性与快速查看

````python
df = df1.copy()   # 使用第1个例子

print("=== 形状与结构 ===")
print(f".shape    → {df.shape}")           # (行数, 列数)
print(f".ndim     → {df.ndim}")            # 2
print(f".size     → {df.size}")            # 行数 × 列数
print(f".columns  → {list(df.columns)}")   # 列名列表
print(f".index    → {list(df.index)}")     # 行索引
print(f".dtypes:\n{df.dtypes}")            # 每列数据类型

print("\n=== 快速查看 ===")
print("head(2):")
print(df.head(2))            # 前 n 行（默认 5）
print("\ntail(2):")
print(df.tail(2))            # 后 n 行

print("\ninfo():")
df.info()                    # 每列类型、非空数量、内存

print("\ndescribe() — 数值列统计:")
print(df.describe())

print("\ndescribe(include='all') — 所有列:")
print(df.describe(include='all'))
````

## 3.3 列选取与行选取

````python
# ===== 列选取 =====
print("单列（返回 Series）:")
print(df['姓名'])

print("\n多列（返回 DataFrame）:")
print(df[['姓名', '成绩']])

# ===== 行选取：loc（标签）=====
print("\n=== loc（按标签）===")
print(df.loc[0])                          # 第 0 行（整数索引作为标签）
print(df.loc[0:2])                        # 标签切片（含右端！）
print(df.loc[0, '姓名'])                  # 单个值
print(df.loc[[0,2], ['姓名','成绩']])     # 多行多列

# ===== 行选取：iloc（位置）=====
print("\n=== iloc（按位置）===")
print(df.iloc[0])                         # 第 0 行
print(df.iloc[-1])                        # 最后一行
print(df.iloc[0:2])                       # 位置切片（不含右端）
print(df.iloc[1:3, 0:2])                  # 行列同时切片

# ===== 布尔索引 =====
print("\n=== 布尔索引 ===")
print(df[df['成绩'] >= 90])               # 成绩 >= 90
print(df[(df['城市'] == '北京') & (df['成绩'] > 80)])  # 多条件

# at / iat —— 单值快速访问（比 loc/iloc 快）
print(f"\ndf.at[0, '姓名']  = {df.at[0, '姓名']}")
print(f"df.iat[0, 0]      = {df.iat[0, 0]}")
````

# 第 5 章：数据清洗

真实数据通常有缺失值、重复值、格式错误等问题。

## 5.1 缺失值处理

````python
# 构造含缺失值的 DataFrame
dirty = pd.DataFrame({
    '姓名':  ['小明', '小红', None,  '小丽', '小明'],
    '年龄':  [18,     np.nan, 20,    18,     18],
    '成绩':  [85,     92,     np.nan, 96,    85],
    '城市':  ['北京', '上海', '广州', np.nan, '北京'],
})
print("原始数据:")
print(dirty)

# ① 检测缺失值
print("\nisnull():")
print(dirty.isnull())

print("\n各列缺失数量:")
print(dirty.isnull().sum())

print(f"\n总缺失数: {dirty.isnull().sum().sum()}")
print(f"任一列含缺失的行:\n{dirty[dirty.isnull().any(axis=1)]}")

# ② 删除缺失值
print("\ndropna()（删除含 NaN 的行）:")
print(dirty.dropna())

print("\ndropna(subset=['成绩'])（只看成绩列）:")
print(dirty.dropna(subset=['成绩']))

print("\ndropna(thresh=3)（至少有 3 个非 NaN 才保留）:")
print(dirty.dropna(thresh=3))

# ③ 填充缺失值
print("\nfillna 填充策略:")
print(dirty['年龄'].fillna(dirty['年龄'].mean()))          # 均值填充
print(dirty['城市'].fillna(method='ffill'))                 # 前向填充（用前一行值）
print(dirty['城市'].fillna(method='bfill'))                 # 后向填充

filled = dirty.fillna({'姓名': '未知', '年龄': dirty['年龄'].mean(), '成绩': 0, '城市': '未知'})
print("\n多列分别填充:")
print(filled)
````

## 5.2 重复值处理

````python
print("检测重复行:")
print(dirty.duplicated())                         # True = 重复行

print(f"\n共有 {dirty.duplicated().sum()} 行重复")

print("\n重复行内容:")
print(dirty[dirty.duplicated(keep=False)])         # keep=False 显示所有重复项

print("\n删除重复行（保留第一次出现）:")
print(dirty.drop_duplicates())

print("\n按姓名列去重（保留最后一次）:")
print(dirty.drop_duplicates(subset=['姓名'], keep='last'))
````

## 5.3 数据类型转换

````python
mixed = pd.DataFrame({
    '学号':  ['001', '002', '003'],
    '成绩':  ['85', '92.5', '78'],      # 字符串形式的数字
    '是否及格': ['True', 'True', 'False'],
    '日期':  ['2024-01-01', '2024-02-15', '2024-03-20'],
})
print("原始类型:")
print(mixed.dtypes)

# astype —— 强制类型转换
mixed['学号'] = mixed['学号'].astype(str)
mixed['成绩'] = mixed['成绩'].astype(float)
mixed['是否及格'] = mixed['是否及格'].map({'True': True, 'False': False})
mixed['日期'] = pd.to_datetime(mixed['日期'])

print("\n转换后类型:")
print(mixed.dtypes)
print(mixed)

# pd.to_numeric —— 含错误字符时更安全
s_num = pd.Series(['1', '2', 'abc', '4'])
print("\n含非数字字符，errors='coerce' 转为 NaN:")
print(pd.to_numeric(s_num, errors='coerce'))

# category 类型 —— 节省内存
df_big = df_csv.copy()
df_big['年级'] = df_big['年级'].astype('category')
print(f"\n年级列改为 category: {df_big['年级'].dtype}")
print(f"类别: {list(df_big['年级'].cat.categories)}")
````

# 第 6 章：数据操作

## 6.1 新增列与修改值

````python
stu = df_csv.copy()

# ① 直接赋值新列
stu['总分'] = stu['语文'] + stu['数学'] + stu['英语']
stu['平均分'] = stu['总分'] / 3

# ② 条件列（apply + lambda）
stu['等级'] = stu['平均分'].apply(lambda x: 'A' if x >= 90 else ('B' if x >= 75 else 'C'))

# ③ np.where / np.select 批量条件赋值
stu['是否优秀'] = np.where(stu['平均分'] >= 85, '优秀', '一般')

# ④ assign —— 链式风格（不修改原 df，返回新 df）
stu2 = stu.assign(
    语数之和 = lambda d: d['语文'] + d['数学'],
    语数差   = lambda d: (d['语文'] - d['数学']).abs()
)

print(stu[['姓名','语文','数学','英语','总分','平均分','等级','是否优秀']].head(8))

# ⑤ 修改特定位置
stu.loc[stu['姓名'] == '学生00', '语文'] = 99
print(f"\n修改后学生00语文成绩: {stu.loc[stu['姓名']=='学生00', '语文'].values}")
````

## 6.2 删除行列 / 重命名 / 排序

````python
base = stu[['姓名','年级','语文','数学','英语','总分','平均分','等级']].copy()

# === 删除 ===
print("删除 '等级' 列:")
print(base.drop(columns=['等级']).head(3))

print("\n删除第 0 行:")
print(base.drop(index=0).head(3))

# === 重命名 ===
renamed = base.rename(columns={'语文': 'Chinese', '数学': 'Math', '英语': 'English'})
print("\n重命名列:")
print(renamed.head(3))

# 重命名索引
renamed2 = base.rename(index={0: 'row_A', 1: 'row_B'})
print("\n重命名行索引:")
print(renamed2.head(3))

# === 排序 ===
print("\n按总分降序:")
print(base.sort_values('总分', ascending=False).head(5))

print("\n多列排序（年级升序，总分降序）:")
print(base.sort_values(['年级','总分'], ascending=[True, False]).head(6))

# 重置索引
reset = base.sort_values('总分', ascending=False).reset_index(drop=True)
print("\n排序后重置索引:")
print(reset.head(4))
````

## 6.3 apply / map / applymap

| 方法 | 作用对象 | 说明 |
|------|----------|------|
| `Series.map()` | 单列 | 值映射或函数变换 |
| `DataFrame.apply()` | 行或列 | 对整行/列应用函数 |
| `DataFrame.map()` | 每个元素（pandas 2.1+，旧版用 applymap）| 元素级函数变换 |

````python
score_df = base[['姓名','语文','数学','英语']].copy()

# Series.map —— 值替换
grade_map = {'高一': 'G1', '高二': 'G2', '高三': 'G3'}
print("年级英文缩写:")
print(base['年级'].map(grade_map).head())

# Series.map —— 函数变换
print("\n语文成绩标准化（0-1）:")
col = score_df['语文']
print(((col - col.min()) / (col.max() - col.min())).round(3).head())

# DataFrame.apply —— 对列（axis=0）
print("\n各科平均分:")
print(score_df[['语文','数学','英语']].apply(np.mean))

print("\n各行最高分:")
print(score_df[['语文','数学','英语']].apply(max, axis=1).head())

# DataFrame.apply —— 返回多值（用 pd.Series）
def score_stats(row):
    scores = row[['语文','数学','英语']]
    return pd.Series({'最高': scores.max(), '最低': scores.min(), '极差': scores.max()-scores.min()})

print("\n每行分数统计:")
print(score_df.apply(score_stats, axis=1).head())

# DataFrame.map（元素级）
num_cols = score_df[['语文','数学','英语']]
print("\n每个分数 +5 的奖励分:")
print(num_cols.map(lambda x: x + 5).head(3))
````

# 第 7 章：索引进阶

## 7.1 设置与重置索引

````python
idx_df = base.copy()

# set_index —— 将某列设为行索引
idx_df = idx_df.set_index('姓名')
print("set_index('姓名'):")
print(idx_df.head(3))

# 用标签索引
print(f"\n学生00的总分: {idx_df.loc['学生00', '总分']}")

# reset_index —— 索引还原为列
print("\nreset_index():")
print(idx_df.reset_index().head(3))
````

## 7.2 多级索引（MultiIndex）

MultiIndex 允许一个轴上有多个层级，常用于层次化数据（班级→学生，年→月→日）。

````python
# 构造多级索引 DataFrame
multi = base.set_index(['年级', '姓名'])
print("多级索引（年级→姓名）:")
print(multi.head(8))

print("\n取高一年级所有学生:")
print(multi.loc['高一'])

print("\n取高一/学生01:")
print(multi.loc[('高一', '学生01')])

# sort_index —— 多级索引需要排序才能高效切片
multi_sorted = multi.sort_index()
print("\nslice 切片（高一到高二）:")
print(multi_sorted.loc['高一':'高二'].head(5))

# swaplevel —— 交换层级
print("\n交换层级（姓名→年级）:")
print(multi_sorted.swaplevel().head(4))

# stack / unstack —— 行列转换
unstacked = multi_sorted['语文'].unstack(level=0)   # 将年级层展开为列
print("\nunstack 后（每列=年级）:")
print(unstacked.head(5))
````

# 第 8 章：统计与分组聚合

## 8.1 基本统计

````python
nums = base[['语文','数学','英语','总分','平均分']]

# 逐列统计
print("各列均值:")
print(nums.mean())

print("\n各列标准差:")
print(nums.std())

print("\n相关系数矩阵:")
print(nums.corr().round(3))

# 频率统计
print("\n各年级人数:")
print(base['年级'].value_counts())

print("\n各等级比例:")
print(base['等级'].value_counts(normalize=True).round(3))

# 累积统计
print("\n总分累积和（前5）:")
print(base['总分'].cumsum().head(5))

# nunique / unique
print(f"\n年级种类数: {base['年级'].nunique()}")
print(f"年级唯一值: {base['年级'].unique()}")
````

## 8.2 groupby 分组聚合

````python
g = base.groupby('年级')

# ① 单个聚合
print("各年级平均分:")
print(g['平均分'].mean().round(2))

print("\n各年级人数:")
print(g.size())

# ② 多个聚合函数（agg）
print("\n各年级语数英统计:")
print(g[['语文','数学','英语']].agg(['mean','max','min']).round(2))

# ③ 对不同列用不同函数
print("\n自定义聚合:")
result = g.agg({
    '语文': ['mean', 'max'],
    '数学': 'mean',
    '总分': ['mean', 'sum']
}).round(2)
print(result)

# ④ 自定义聚合函数
print("\n各年级语文极差（max-min）:")
print(g['语文'].agg(lambda x: x.max() - x.min()))

# ⑤ transform —— 保持原索引，聚合值广播回原行
base['年级平均分'] = g['平均分'].transform('mean')
print("\ntransform 后（年级平均分列广播）:")
print(base[['姓名','年级','平均分','年级平均分']].head(8))

# ⑥ filter —— 保留满足条件的分组
print("\n保留平均分 > 75 的年级:")
print(g.filter(lambda x: x['平均分'].mean() > 75)[['姓名','年级','平均分']].head(6))

# ⑦ 多列分组
print("\n按年级+等级分组计数:")
print(base.groupby(['年级','等级']).size().unstack(fill_value=0))
````

## 8.3 透视表 pivot_table

````python
# pivot_table：行=年级，列=等级，值=平均分，聚合函数=均值
pt = pd.pivot_table(
    base,
    values='平均分',
    index='年级',
    columns='等级',
    aggfunc='mean',
    fill_value=0,
    margins=True,     # 添加行列汇总
    margins_name='全部'
)
print("透视表（各年级各等级平均分）:")
print(pt.round(2))

# crosstab —— 交叉频率表
print("\n交叉频次表（年级 × 等级）:")
print(pd.crosstab(base['年级'], base['等级'], margins=True))
````

# 第 9 章：数据合并

## 9.1 concat —— 堆叠拼接

````python
df_a = pd.DataFrame({'姓名':['小明','小红'], '语文':[85,92], '数学':[90,88]})
df_b = pd.DataFrame({'姓名':['小华','小丽'], '语文':[78,96], '数学':[82,95]})
df_c = pd.DataFrame({'姓名':['小华','小明'], '英语':[88,79], '物理':[91,85]})

print("=== 纵向拼接（axis=0，行方向）===")
print(pd.concat([df_a, df_b], ignore_index=True))

print("\n=== 横向拼接（axis=1，列方向）===")
print(pd.concat([df_a, df_c], axis=1))

print("\n=== join='outer'（保留所有列，缺失填 NaN）===")
df_d = pd.DataFrame({'姓名':['小明','小红'], '历史':[88,75]})
print(pd.concat([df_a, df_d], join='outer', ignore_index=True))

print("\n=== join='inner'（只保留共同列）===")
print(pd.concat([df_a, df_d], join='inner', ignore_index=True))

print("\n=== keys 参数（添加外层标签区分来源）===")
print(pd.concat([df_a, df_b], keys=['班级A', '班级B']))
````

## 9.2 merge —— 按键列关联合并（类似 SQL JOIN）

| how | 说明 |
|-----|------|
| `'inner'` | 两表共有的键（默认） |
| `'left'` | 保留左表所有行 |
| `'right'` | 保留右表所有行 |
| `'outer'` | 保留两表所有行 |

````python
students = pd.DataFrame({
    '学号': [1, 2, 3, 4],
    '姓名': ['小明', '小红', '小华', '小丽'],
    '班级': ['A', 'A', 'B', 'B']
})
scores = pd.DataFrame({
    '学号': [1, 2, 3, 5],      # 5 不在 students 中，4 没有成绩
    '语文': [85, 92, 78, 88],
    '数学': [90, 88, 82, 91]
})

print("students:")
print(students)
print("\nscores:")
print(scores)

print("\n=== inner join（默认）===")
print(pd.merge(students, scores, on='学号', how='inner'))

print("\n=== left join（保留左表所有）===")
print(pd.merge(students, scores, on='学号', how='left'))

print("\n=== right join ===")
print(pd.merge(students, scores, on='学号', how='right'))

print("\n=== outer join ===")
print(pd.merge(students, scores, on='学号', how='outer'))

# 左右列名不同时
scores2 = scores.rename(columns={'学号': 'sid'})
print("\n=== 左右键名不同（left_on/right_on）===")
print(pd.merge(students, scores2, left_on='学号', right_on='sid', how='inner').drop('sid', axis=1))

# 同名但非键列加后缀
info = pd.DataFrame({'学号':[1,2,3], '姓名':['M','R','H'], '班级':['A','A','B']})
print("\n=== 同名非键列加后缀 ===")
print(pd.merge(students, info, on='学号', suffixes=('_left', '_right')))
````

# 第 10 章：时间序列

## 10.1 日期创建与属性

````python
# 创建日期序列
dates = pd.date_range('2024-01-01', periods=12, freq='ME')   # 每月最后一天
print("月末日期序列:")
print(dates)

# 日期属性
ts = pd.Timestamp('2024-06-15 14:30:00')
print(f"\nTimestamp 属性:")
print(f"  年={ts.year}  月={ts.month}  日={ts.day}")
print(f"  时={ts.hour}  分={ts.minute}  秒={ts.second}")
print(f"  星期几={ts.day_name()}  季度={ts.quarter}")
print(f"  是否月末={ts.is_month_end}")

# 构造时间序列 DataFrame
np.random.seed(42)
ts_df = pd.DataFrame({
    '日期': pd.date_range('2024-01-01', periods=365, freq='D'),
    '销售额': np.random.randint(800, 2000, 365) + np.sin(np.arange(365)/30*np.pi)*300
})
ts_df = ts_df.set_index('日期')

# dt accessor —— 从列中提取日期属性
ts_df2 = ts_df.reset_index()
ts_df2['月'] = ts_df2['日期'].dt.month
ts_df2['星期'] = ts_df2['日期'].dt.day_name()
ts_df2['季度'] = ts_df2['日期'].dt.quarter
print("\ndt accessor 提取:")
print(ts_df2.head(3))
````

## 10.2 时间切片与重采样

````python
# 时间切片
print("2024-03 月数据:")
print(ts_df.loc['2024-03'].head(5))

print("\n2024-Q1 数据（前3行）:")
print(ts_df.loc['2024-01':'2024-03'].head(3))

# resample —— 重采样（降采样：日→周/月）
monthly = ts_df.resample('ME').agg({'销售额': ['sum', 'mean', 'max']})
monthly.columns = ['月销售额', '日均销售额', '最高日销售额']
print("\n月度汇总:")
print(monthly.round(0))

# rolling —— 滚动窗口（移动平均）
ts_df['7日均值'] = ts_df['销售额'].rolling(window=7, min_periods=1).mean()
ts_df['30日均值'] = ts_df['销售额'].rolling(window=30, min_periods=1).mean()

# shift —— 时间移位（做同比/环比）
ts_monthly = ts_df.resample('ME')['销售额'].sum().reset_index()
ts_monthly['上月销售额'] = ts_monthly['销售额'].shift(1)
ts_monthly['环比增长率'] = (ts_monthly['销售额'] / ts_monthly['上月销售额'] - 1).round(3)
print("\n月度环比:")
print(ts_monthly.dropna())

# 可视化
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 7))

ax1.plot(ts_df.index, ts_df['销售额'], alpha=0.3, color='steelblue', label='日销售额')
ax1.plot(ts_df.index, ts_df['7日均值'], color='tomato', label='7日均线')
ax1.plot(ts_df.index, ts_df['30日均值'], color='orange', linewidth=2, label='30日均线')
ax1.set_title('2024年日销售额与移动平均')
ax1.legend()

ax2.bar(ts_monthly['日期'], ts_monthly['销售额'], color='steelblue', alpha=0.7)
ax2.set_title('月度销售总额')
ax2.set_ylabel('销售额')

plt.tight_layout()
plt.show()
````

# 第 13 章：综合实战 —— 学生成绩完整分析

将前面所有知识综合运用，完成一份完整的数据分析报告。

````python
np.random.seed(0)
n_stu = 120
subjects = ['语文','数学','英语','物理','化学','历史']

# 模拟真实场景：不同年级基础分不同
grade_base = {'高一': 70, '高二': 75, '高三': 80}
grades = np.random.choice(['高一','高二','高三'], n_stu, p=[0.4, 0.35, 0.25])

raw_data = {'学号': [f'STU{i:03d}' for i in range(1, n_stu+1)], '年级': grades}
for subj in subjects:
    raw_data[subj] = [
        min(100, max(0, int(grade_base[g] + np.random.randn()*12)))
        for g in grades
    ]

# 注入缺失值和重复值
raw_df = pd.DataFrame(raw_data)
raw_df.loc[np.random.choice(n_stu, 8, replace=False), '数学'] = np.nan
raw_df.loc[np.random.choice(n_stu, 5, replace=False), '英语'] = np.nan
raw_df = pd.concat([raw_df, raw_df.iloc[:3]], ignore_index=True)   # 添加 3 条重复

print(f"原始数据形状: {raw_df.shape}")
print(f"缺失值:\n{raw_df.isnull().sum()}")
print(f"重复行: {raw_df.duplicated().sum()}")
````

````python
# ===== 步骤 1：数据清洗 =====
clean = raw_df.drop_duplicates().copy()
for subj in subjects:
    # 各年级分别用本年级均值填充（更合理）
    clean[subj] = clean.groupby('年级')[subj].transform(lambda x: x.fillna(x.mean()))

print(f"清洗后形状: {clean.shape}")
print(f"剩余缺失值: {clean.isnull().sum().sum()}")

# ===== 步骤 2：特征工程 =====
clean['总分'] = clean[subjects].sum(axis=1)
clean['平均分'] = clean['总分'] / len(subjects)
clean['年级排名'] = clean.groupby('年级')['总分'].rank(ascending=False, method='min').astype(int)
clean['全校排名'] = clean['总分'].rank(ascending=False, method='min').astype(int)

conditions = [
    clean['平均分'] >= 90,
    clean['平均分'] >= 80,
    clean['平均分'] >= 70,
    clean['平均分'] >= 60,
]
choices = ['A', 'B', 'C', 'D']
clean['等级'] = np.select(conditions, choices, default='E')

print("\n清洗 + 特征工程后:")
print(clean[['学号','年级','总分','平均分','年级排名','全校排名','等级']].head(8))
````

````python
# ===== 步骤 3：分析与可视化 =====
fig = plt.figure(figsize=(16, 12))

# 图1：各年级各科平均分
ax1 = fig.add_subplot(2, 3, 1)
grade_avg = clean.groupby('年级')[subjects].mean().round(1)
grade_avg.T.plot(kind='bar', ax=ax1, rot=30, colormap='Set2')
ax1.set_title('各年级各科平均分')
ax1.set_ylabel('分数')
ax1.legend(fontsize=8)

# 图2：全校成绩分布直方图
ax2 = fig.add_subplot(2, 3, 2)
clean['平均分'].plot(kind='hist', bins=15, ax=ax2, color='steelblue',
                     edgecolor='k', alpha=0.8)
ax2.axvline(clean['平均分'].mean(), color='red', linestyle='--',
            label=f'均值 {clean["平均分"].mean():.1f}')
ax2.set_title('全校平均分分布')
ax2.set_xlabel('平均分')
ax2.legend()

# 图3：各年级等级分布堆叠条形图
ax3 = fig.add_subplot(2, 3, 3)
grade_level = pd.crosstab(clean['年级'], clean['等级'])
grade_level_pct = grade_level.div(grade_level.sum(axis=1), axis=0) * 100
grade_level_pct.plot(kind='bar', stacked=True, ax=ax3, colormap='RdYlGn', rot=0)
ax3.set_title('各年级等级分布（%）')
ax3.set_ylabel('占比 %')
ax3.legend(title='等级', fontsize=8, loc='lower right')

# 图4：六科箱线图
ax4 = fig.add_subplot(2, 3, 4)
clean[subjects].plot(kind='box', ax=ax4, patch_artist=True)
ax4.set_title('六科成绩分布（箱线图）')
ax4.set_ylabel('分数')

# 图5：语文 vs 数学散点图（按年级着色）
ax5 = fig.add_subplot(2, 3, 5)
colors = {'高一': 'steelblue', '高二': 'tomato', '高三': 'green'}
for grade, grp in clean.groupby('年级'):
    ax5.scatter(grp['语文'], grp['数学'], c=colors[grade],
                label=grade, alpha=0.6, s=30)
ax5.set_xlabel('语文')
ax5.set_ylabel('数学')
ax5.set_title('语文 vs 数学（按年级）')
ax5.legend()

# 图6：Top10 全校排名学生总分
ax6 = fig.add_subplot(2, 3, 6)
top10 = clean.nsmallest(10, '全校排名')[['学号','总分','年级']].reset_index(drop=True)
bars = ax6.barh(top10['学号'], top10['总分'],
                color=[colors[g] for g in top10['年级']], alpha=0.85)
ax6.set_xlabel('总分')
ax6.set_title('全校 Top 10 学生总分')
ax6.invert_yaxis()

plt.suptitle('学生成绩数据分析报告', fontsize=15, fontweight='bold')
plt.tight_layout()
plt.show()
````

````python
# ===== 步骤 4：输出报告摘要 =====
print("=" * 55)
print("         学生成绩分析报告摘要")
print("=" * 55)
print(f"\n总学生数: {len(clean)} 人")
print(f"年级分布:\n{clean['年级'].value_counts().to_string()}")

print(f"\n全校平均分: {clean['平均分'].mean():.2f}")
print(f"最高总分:   {clean['总分'].max()} 分（学号: {clean.loc[clean['总分'].idxmax(), '学号']}）")
print(f"最低总分:   {clean['总分'].min()} 分")

print(f"\n各年级平均分:")
print(clean.groupby('年级')['平均分'].mean().round(2).to_string())

print(f"\n各等级人数:")
print(clean['等级'].value_counts().sort_index().to_string())

print(f"\n各科难度排名（均分从低到高）:")
subj_avg = clean[subjects].mean().sort_values()
for i, (subj, avg) in enumerate(subj_avg.items(), 1):
    print(f"  {i}. {subj}: {avg:.2f}")

print(f"\n科目间相关性（最强的 3 对）:")
corr_mat = clean[subjects].corr()
corr_pairs = corr_mat.where(np.triu(np.ones_like(corr_mat, dtype=bool), k=1))
corr_series = corr_pairs.stack().sort_values(ascending=False)
for (s1, s2), val in corr_series.head(3).items():
    print(f"  {s1} ↔ {s2}: {val:.3f}")
print("=" * 55)
````

## B. 常见陷阱

| 陷阱 | 正确做法 |
|------|----------|
| `df[0:2]` 切行用位置 | 用 `df.iloc[0:2]`（明确） |
| `df.loc[0:2]` 是标签切片，**含右端 2** | 注意与 iloc 的区别 |
| 链式赋值 `df[df.x>0]['y'] = 1` | 改用 `df.loc[df.x>0, 'y'] = 1` |
| `fillna(method=...)` Pandas 2.2 弃用 | 改用 `ffill()` / `bfill()` 方法 |
| 忘记 `copy()` 导致修改原 df | 赋值新变量前加 `.copy()` |
| 读取中文 CSV 乱码 | `encoding='utf-8-sig'` 或 `gbk` |
