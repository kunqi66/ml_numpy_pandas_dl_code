## series创建

```python
s = pd.Series([2,3,4,5,6]) # series是以列为单位的，必须存储相同类型数据
print(s)
# 可设置索引
s = pd.Series([2,3,4,5,6],index = [1,2,3,4,5])
# 可设置name
s = pd.Series([2,3,4,5,6],index = [1,2,3,4,5], name='number')
print(s)
# 通过字典来创建
s = pd.Series({'a':12,'b':2,'d':4})
print(s)
# 用索引来创建
s = pd.Series(s,index=['a','b'])
print(s)
```
