## series创建

```python
s = pd.Series([2,3,4,5,6]) # series是以列为单位的，必须存储相同类型数据
print(s)
# 可设置索引
s = pd.Series([2,3,4,5,6],index = [1,2,3,4,5])
# 可设置name
s = pd.Series([2,3,4,5,6],index = [1,2,3,4,5], name='number')
print(s)
# 通过字典来创建
s = pd.Series({'a':12,'b':2,'d':4})
print(s)
# 用索引来创建
s = pd.Series(s,index=['a','b'])
print(s)
```

## serise的属性

![](./img/ser属性.png)

```python
# index : Series 的索引对象
print(s.index)
print(s.values)
print(s.shape,s.ndim,s.size)
s.name = 'test'
print(s.dtype,s.name)
print(s.loc['a']) # 显式索引
print(s.iloc[0])  #隐式索引
print(s.loc['a':'b'])  #使用at不支持切片
```

## series访问数据

```python
print(s['a'])
print(s[s<5])
print(s.head()) #前五行
print(s.tail()) # 后五行
```

```python
s = pd.Series([10.2,np.nan,None,3,4,5,6],index=['A','B','C','D','E','F','G'])
print(s)
s.describe()
print(s.head(3))
print(s.head(6))
print(s.tail(3))
```

```python
# 获取元素数量，忽略缺失值
print(s.count())
# 获取索引
print(s.keys())
print(s.index)
# 每个位置是否为某值
s.isin([4,5,6])
print(s.describe())
```
