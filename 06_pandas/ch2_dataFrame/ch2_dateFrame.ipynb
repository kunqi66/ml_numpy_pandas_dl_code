## DateFrame
行为记录，列为series，没列的数据类型需要相同 

```python
import pandas as pd
import numpy as np
```

### dataframe的创建

```python
s1 = pd.Series([1,2,3,4,5])
s2 = pd.Series([6,7,8,9,10])
df = pd.DataFrame({'第一列':s1,'第二列':s2})
print(type(df))
print(type(df['第一列']))
# 使用字典创建DateFrame
# 可通过index 和 colume来调整行列的顺序
df = pd.DataFrame({
    'id':[1,2,3,4,5],
    'name':['tom','jack','alice','bob','allen'],
    'age':[15,17,20,26,30],
    'score':[60.5,80,30,70,83.5]
})
print(df)
```

### DateFrame的属性

![](./img/dataframe属性.png)

```python
# 获取多列数据
print(df[['name','age']])
print(df.head(2))
print(df.tail(3))
# 布尔索引
print(df[(df.score <80) & (df.age>20)])
```
